```python
# url収集
content_list = []
for url in url_list:
# for url in url_list[len(content_list):]:
    try:
        print(url)
        hdr = {'User-Agent': 'Mozilla/5.0'}
        # https://stackoverflow.com/questions/9942594/unicodeencodeerror-ascii-codec-cant-encode-character-u-xa0-in-position-20/9942885
        req = Request(url.encode('ascii', 'ignore').decode('ascii'),headers=hdr)
        html = urlopen(req)
        bsObj = BeautifulSoup(html,"lxml")

        title_obj = bsObj.find("h1",{"class":"articleHeader"})

        article_obj = bsObj.find("div",{"class":"arial_14"})
        date_obj = bsObj.find("div",{"class":"contentSectionDetails"}).find("span")


#         print(title_obj.text)
#         print(article_obj.text)
        print(date_obj.text)
        content_list.append([date_obj.text, title_obj.text, article_obj.text])
    except HTTPError as e:
        print("error : {0}".format(e))
        continue
#     break
```
